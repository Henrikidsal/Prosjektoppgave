```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import ast
```

```python
df = pd.read_csv('PLOTS.csv')
```

## Full size problem default time

```python
# Initialize lists for feasible and non-feasible values
all_values = []
feasible_values = []

# Consider the xxx rows of the dataframe
custom_range_df = df.iloc[0:50]

#display(custom_range_df)


# Loop through each row in the dataframe
for index, row in custom_range_df.iterrows():
    cleaned_row_ofv = row[" Objective Function Values"].replace("np.float64", "")
    cleaned_row_fs = row[" feasible solutions"].replace("np.float64", "")
    try:
        # Parse the cleaned row as a Python literal
        row_updated_ofv = ast.literal_eval(cleaned_row_ofv)
        float_list_ofv = [float(x) for x in row_updated_ofv]

        row_updated_fs = ast.literal_eval(cleaned_row_fs)
        float_list_fs = [float(x) for x in row_updated_fs]

        # Append the parsed row to the list of feasible or non-feasible values
        all_values.extend(float_list_ofv)
        feasible_values.extend(float_list_fs)

        
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing row {index}: {row} - {e}")

# Display the results
print("Feasible values:", feasible_values)
print("all values:", all_values)
print(f"Shape of feasible values: {len(feasible_values)}")
print(f"Shape of all values: {len(all_values)}")

# Create bins
bin_edges = 10 ** np.linspace(4, 7.8, 120)

# Calculate counts for all solutions and feasible solutions
all_counts, _ = np.histogram(all_values, bins=bin_edges)
feasible_counts, _ = np.histogram(feasible_values, bins=bin_edges)

# Subtract feasible counts from total counts to get the counts of infeasible solutions
infeasible_counts = all_counts - feasible_counts

# Plot infeasible (blue) and feasible (green) solutions stacked
plt.bar(bin_edges[:-1], all_counts, width=np.diff(bin_edges), align='edge', color='blue', edgecolor='blue', label='All solutions', alpha=0.6)
plt.bar(bin_edges[:-1], feasible_counts, width=np.diff(bin_edges), align='edge', color='green', edgecolor='green', label='Feasible solutions', alpha=0.8)

# Add Gurobi optimal solution line
plt.axvline(966987, color='red', linestyle='dashed', linewidth=1, label='Gurobi optimal solution')

# Customize the plot
plt.title('Full size Problem')
plt.xlim([800000, 5e+7])
plt.xlabel('Objective function value', fontsize=15)
plt.ylabel('Number of solutions', fontsize=15)
plt.xscale('log')
plt.yscale('log')
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=8, loc='upper left')
plt.savefig('plots/full_size_problem.png')

# Show the plot
plt.show()
```

```python
import ast

# Initialize the lists to store feasible and not feasible values
feasible_values = []
not_feasible_values = []

# Loop through the rows from index 52 to 101 (adjusting for zero-based indexing)
for index in range(340, 390):  # Loop over rows 52 to 101
    try:
        # Access and clean the "Objective Function Values" column
        objective_row = df[" Objective Function Values"].iloc[index].replace("np.float64", "")
        # Convert the cleaned string to a list of floats
        objective_values = [float(x) for x in ast.literal_eval(objective_row)]
        
        # Access and clean the "feasible solutions" column
        feasible_row = df[" feasible solutions"].iloc[index].replace("np.float64", "")
        # Convert the cleaned string to a list of floats
        feasible_solutions = [float(x) for x in ast.literal_eval(feasible_row)]
        
        # Classify values into feasible and not feasible
        for value in objective_values:
            if value in feasible_solutions:
                feasible_values.append(value)
            else:
                not_feasible_values.append(value)

    except (ValueError, SyntaxError) as e:
        print(f"Error parsing row {index}: {e}")

# Print the results
print("Feasible Values:", feasible_values)
print("Not Feasible Values:", not_feasible_values)


new_df = df.iloc[310:320]
display(new_df.head(20))

custom_anealing_time_values = []

for index, row in enumerate(new_df[" feasible solutions"]):
    # Replace 'np.float64' with nothing to clean the string
    cleaned_row = row.replace("np.float64", "")
    try:
        # Safely evaluate the cleaned string
        row_updated = ast.literal_eval(cleaned_row)
        # Convert to regular floats
        float_list = [float(x) for x in row_updated]
        custom_anealing_time_values.extend(float_list)  # Add the values to the final list
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing row {index}: {row} - {e}")
print("Custom Annealing Time Values:", custom_anealing_time_values)
```
