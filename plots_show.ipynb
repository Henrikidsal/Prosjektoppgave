```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import ast
```

```python
df = pd.read_csv('PLOTS.csv')
```

## Full size problem default time

```python
# Initialize lists for feasible and non-feasible values
all_values = []
feasible_values = []

# Consider the xxx rows of the dataframe
custom_range_df = df.iloc[0:50]

#display(custom_range_df)


# Loop through each row in the dataframe
for index, row in custom_range_df.iterrows():
    cleaned_row_ofv = row[" Objective Function Values"].replace("np.float64", "")
    cleaned_row_fs = row[" feasible solutions"].replace("np.float64", "")
    try:
        # Parse the cleaned row as a Python literal
        row_updated_ofv = ast.literal_eval(cleaned_row_ofv)
        float_list_ofv = [float(x) for x in row_updated_ofv]

        row_updated_fs = ast.literal_eval(cleaned_row_fs)
        float_list_fs = [float(x) for x in row_updated_fs]

        # Append the parsed row to the list of feasible or non-feasible values
        all_values.extend(float_list_ofv)
        feasible_values.extend(float_list_fs)

        
    except (ValueError, SyntaxError) as e:
        print(f"Error parsing row {index}: {row} - {e}")

# Display the results
print("Feasible values:", feasible_values)
print("all values:", all_values)
print(f"Shape of feasible values: {len(feasible_values)}")
print(f"Shape of all values: {len(all_values)}")

# Adjust the bin range to start from the best feasible solution
bin_edges = 10 ** np.linspace(4, 7.8, 120)

# Calculate counts for all solutions and feasible solutions
all_counts, _ = np.histogram(all_values, bins=bin_edges)
feasible_counts, _ = np.histogram(feasible_values, bins=bin_edges)

# Subtract feasible counts from total counts to get the counts of infeasible solutions
infeasible_counts = all_counts - feasible_counts

# Plot infeasible (blue) and feasible (green) solutions stacked
plt.bar(bin_edges[:-1], all_counts, width=np.diff(bin_edges), align='edge', edgecolor='blue', label='All solutions')
plt.bar(bin_edges[:-1], feasible_counts, width=np.diff(bin_edges), align='edge', color='green', edgecolor='green', label='Feasible solutions')

# Add Gurobi optimal solution line
plt.axvline(966987, color='red', linestyle='dashed', linewidth=1, label='Gurobi optimal solution')

# Customize the plot
plt.title('Full size Problem', fontsize=11)
plt.xlim([800000, 5e+7])
plt.ylim([0.5, 2e+3])
plt.xlabel('Objective function value', fontsize=11)
plt.ylabel('Number of solutions', fontsize=11)
plt.xscale('log')
plt.yscale('log')
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=8, loc='upper left')
plt.savefig('plots/full_size_problem.png')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt

df = pd.read_csv('plotting2.csv')

# Convert string representations of lists into actual Python lists
df['reserves'] = df['reserves'].apply(ast.literal_eval)
df['demand'] = df['demand'].apply(ast.literal_eval)
df['output_thermal_gens'] = df['output_thermal_gens'].apply(ast.literal_eval)
df['output_renewable_gens'] = df['output_renewable_gens'].apply(ast.literal_eval)
df['output_spinning_reserves'] = df['output_spinning_reserves'].apply(ast.literal_eval)

# Extract data arrays
reserve = df['reserves'][0]
demand = df['demand'][0]
pg = df['output_thermal_gens'][0]
pw = df['output_renewable_gens'][0]
rg = df['output_spinning_reserves'][0]

# Convert them to numpy arrays for convenience
pg = np.array(pg)
pw = np.array(pw)
rg = np.array(rg)
demand = np.array(demand)
reserve = np.array(reserve)

# Number of elements
indices = np.arange(len(pg))

# Bar width
bar_width = 0.5

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot stacked positive bars for pg, pw, and rg
bar1 = ax.bar(indices, pg, bar_width, label="Thermal generators", color="blue")
bar2 = ax.bar(indices, pw, bar_width, bottom=pg, label="Renewable generators", color="orange")
bar3 = ax.bar(indices, rg, bar_width, bottom=pg+pw, label="Spinning reserves", color="green")

# Plot negative bars for demand and reserve
bar4 = ax.bar(indices, -demand, bar_width, label="Demand", color="red")
bar5 = ax.bar(indices, -reserve, bar_width, bottom=-demand, label="Reserve", color="gray")

# Optional plot line (adjust if needed)
plot1 = ax.plot(indices, pg + pw, ls="--", label="Demand", color="red")

# Adjust y-ticks to show absolute values
ax.set_yticks(ax.get_yticks())
ax.set_yticklabels([f"{abs(y):.0f}" for y in ax.get_yticks()])

# Add labels, legend, and formatting
ax.set_xlabel("Time (h)", fontsize=12)
ax.set_ylabel("Power (MW)", fontsize=12)
ax.legend(loc="center left", bbox_to_anchor=(1, 0.5), fontsize=10)
ax.grid(True, axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.ylim([-5000, 5000])
plt.show()
```
